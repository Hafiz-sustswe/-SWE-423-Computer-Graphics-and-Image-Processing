# Bresenham Line Drawing Algorithm

This notebook demonstrates the implementation of the Bresenham Line Drawing Algorithm. This efficient algorithm is used for drawing lines on a discrete grid, such as a pixelated display. It minimizes the use of floating-point arithmetic, using decision parameters to incrementally choose the pixels that best approximate the desired line.

```python
# Importing the required library
import matplotlib.pyplot as plt
```

```python
def visual_show(all_points, point1, point2):
    """
    Visualizes the line between two points using Bresenham's algorithm.
    
    Parameters:
    - all_points: List of tuples representing the points (pixels) on the line.
    - point1, point2: The starting and ending point of the line.
    """
    # Extract x and y coordinates from list
    x_coord = [coord[0] for coord in all_points]
    y_coord = [coord[1] for coord in all_points]
    # Plot the calculated points
    plt.plot(x_coord, y_coord, marker='o', color='red')
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title(f"Line Between {point1} and {point2}")
    plt.grid(True)
    # Highlight the start and end points
    x = [point1[0], point2[0]]
    y = [point1[1], point2[1]]
    plt.plot(x, y, marker='o', color='blue')
    plt.show()
```

Handles the drawing of a line where the absolute slope is less than one.

```python
def slope_less_than_one(point1, point2):
    """
    Handles the drawing of a line where the absolute slope is less than one.
    """
    print('Points are: ', point1, point2)
    print('Slope is lesser than one')
    all_points = []
    all_p = []
    x1, y1 = point1
    x2, y2 = point2
    x, y = x1, y1
    dy = y2 - y1
    dx = x2 - x1
    p = 2 * dy - dx

    while x <= x2:
        all_points.append((x, y))
        all_p.append(p)
        x += 1
        if p < 0:
            p = p + 2 * dy
        else:
            p = p + 2 * dy - 2 * dx
            y += 1
    print("Intermediate Points are: ", all_points)
    print("Decision Parameters are: ", all_p)
    visual_show(all_points, point1, point2)
```

Handles the drawing of a line where the absolute slope is greater than or equal to one.

```python
def slope_greater_than_one(point1, point2):
    """
    Handles the drawing of a line where the absolute slope is greater than or equal to one.
    """
    print('Points are: ', point1, point2)
    print('Slope is greater or equal to one')
    all_points = []
    all_p = []
    x1, y1 = point1
    x2, y2 = point2
    x, y = x1, y1
    dy = y2 - y1
    dx = x2 - x1
    p = 2 * dx - dy

    while y <= y2:
        all_points.append((x, y))
        all_p.append(p)
        y += 1
        if p < 0:
            p = p + 2 * dx
        else:
            p = p + 2 * dx - 2 * dy
            x += 1
    print("Intermediate Points are: ", all_points)
    print("Decision Parameters are: ", all_p)
    visual_show(all_points, point1, point2)
```

The Breshman algorithm

```python
def bresenham_line(point1, point2):
    """
    Determines the primary function to use based on the slope of the line.
    """
    x1, y1 = point1
    x2, y2 = point2
    dy = y2 - y1
    dx = x2 - x1
    if abs(dy) < abs(dx):
        # Slope < 1
        if x1 > x2:
            slope_less_than_one(point2, point1)
        else:
            slope_less_than_one(point1, point2)
    else:
        # Slope >= 1
        if y1 > y2:
            slope_greater_than_one(point2, point1)
        else:
            slope_greater_than_one(point1, point2)
```

```python
def main():
    print('\nBresenham Line Drawing Algorithm')
    # Example test cases
    point1 = [1, 1]
    point2 = [8, 4]
    bresenham_line(point1, point2)

    point1 = [1, 1]
    point2 = [4, 8]
    bresenham_line(point1, point2)

# Call the main function to execute the test cases
if __name__ == "__main__":
    main()
```
